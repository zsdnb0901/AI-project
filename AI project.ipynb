# Machine Learning area

```python
# seperate the features and labels
label = df3[df3.columns[5]].values
df_feature = df3.drop(["ARREST"],axis=1,inplace=False)
feature = df_feature.values

# min max scaler
from sklearn.preprocessing import MinMaxScaler
scalered_number = MinMaxScaler(feature_range=(0,1))
scal_feature = scalered_number.fit_transform(feature)

# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scal_feature,label,random_state = 2,test_size=0.2)
```

```python
# import the packages
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
```

**Random Forest**

```python
# this is based on random forest
_rf_model = RandomForestClassifier(n_estimators=40)
_rf_model.fit(x_train,y_train)
y_predict_rf = _rf_model.predict(x_test)
rf_predict_score = _rf_model.score(x_test,y_test)

print("The confusion matrix of Random Forest:\n",confusion_matrix(y_test,y_predict_rf))
print("The score of Logistic Regression:\n",(rf_predict_score*100).round(2))
```

**Analysis on random forest**

```python
# for testing how n_estimators affect the modle, I will do the analysis here
scores_1 = []
for k in range(40,61):
    _rf_model = RandomForestClassifier(n_estimators=k)
    _rf_model.fit(x_train,y_train)
    y_predict_rf = _rf_model.predict(x_test)
    rf_predict_score = _rf_model.score(x_test,y_test)
    scores_1.append(rf_predict_score)
```

```python
scores_1 = np.array(scores_1)
plt.figure(figsize = (8,6))
plt.suptitle("Random forest n_estimators vs accuracy")
plt.plot(scores_1,color='red', marker='o', markersize=8)
plt.xticks(range(0,21), range(40,61))
plt.show()
print("The highest score for random forest:", scores_1.max().round(2)*100,"%")
```

**Base on this plot, we know that we have the highest scores as around 90%  when n_estimator equals to 53**

**Logistic Regression**

```python
_lr_model = LogisticRegression()
_lr_model.fit(x_train,y_train)
# y_head_lr = lr_model.predict(x_test)
y_predict_lr = _lr_model.predict(x_test)
lr_predict_score = _lr_model.score(x_test,y_test)

print("The confusion matrix of Logistic Regression:\n",confusion_matrix(y_test,y_predict_lr))
print("The score of Logistic Regression:\n",(lr_predict_score*100).round(2))
```

**Adaboost**

```python
_ada_model = AdaBoostClassifier(n_estimators=100)
_ada_model.fit(x_train,y_train)
y_predict_ada = _ada_model.predict(x_test)
ada_predict_score = _ada_model.score(x_test,y_test)

print("The confusion matrix of Ababoost:\n",confusion_matrix(y_test,y_predict_ada))
print("The score of Adaboost:\n",(ada_predict_score*100).round(2))
```

**Analysis on adaboost**

```python
# for testing how n_estimators affect the modle, I will do the analysis here

scores_2 = []
for i in range(100,121):
    _ada_model = AdaBoostClassifier(n_estimators=i)
    _ada_model.fit(x_train,y_train)
    y_predict_ada = _ada_model.predict(x_test)
    ada_predict_score = _ada_model.score(x_test,y_test)
    scores_2.append(ada_predict_score)
```

```python
scores_2 = np.array(scores_2)
plt.figure(figsize = (10,9))
plt.suptitle("Adaboost n_estimators vs accuracy")
plt.plot(scores_2,color='red', marker='o', markersize=8)
plt.xticks(range(0,21), range(100,121))
plt.show()
plt.show(print("The highest score for Adaboost:", scores_2.max().round(2)*100,"%"))
```

**Base on this plot, we know that we have the highest scores as around 88%  when n_estimator equals to 119**

**Majority Voting**

```python
from sklearn.ensemble import VotingClassifier
vc_model = VotingClassifier(estimators=[('rf', _rf_model), ('logistic', _lr_model), ('ada', _ada_model)], 
                            voting='hard')
vc_model.fit(x_train,y_train)
y_predict_vc = vc_model.predict(x_test)
vc_predict_score = vc_model.score(x_test,y_test)

print("The confusion matrix of vc:\n",confusion_matrix(y_test,y_predict_vc))
print("The score of vc:\n",(vc_predict_score*100).round(2))
```
